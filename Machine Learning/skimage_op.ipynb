# skimage

## 读取图像

````python
import skimage.io as io
image_filename = '../res/skimage_ico.png'
image = io.imread(image_filename)
io.imshow(image)
````

## 图像扩展 镜像反转复制
``` python
np.pad(array, pad_width, mode, **kwargs)
```

````python
import numpy as np
padd = 80
imga = np.pad(np.reshape(image,newshape=[-1,image.shape[0],image.shape[1],image.shape[2]]), # 先对图像reshape
              [[0, 0], [padd, padd], [padd, padd], [0, 0]], mode='reflect')
imga = np.reshape(imga, newshape=[imga.shape[1],imga.shape[2],imga.shape[3]])
io.imshow(imga)
# io.imsave('./res/skimage_ico_pad.png',imga)
````

## 图像倍数复制

````python
padd = 2 #倍數
imgdou = np.zeros([image.shape[0]*padd, image.shape[1]*padd,3])
for jj in range(padd):
    for kk in range(padd):
        imgdou[jj*image.shape[0]:(jj+1)*image.shape[0],kk*image.shape[1]:(kk+1)*image.shape[1],:] = np.uint8(image[:,:,:])
imgdou = np.uint8(imgdou)
io.imshow(imgdou)
````
